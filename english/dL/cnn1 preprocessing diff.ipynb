```
# Explore For Noise
Data['tweet']=Data['tweet'].apply(nfx.remove_stopwords)

Data['tweet']=Data['tweet'].apply(nfx.remove_emojis)
Data['tweet']=Data['tweet'].apply(nfx.remove_html_tags)

Data['tweet']=Data['tweet'].apply(nfx.remove_numbers)
Data['tweet']=Data['tweet'].apply(nfx.remove_punctuations)
Data['tweet']=Data['tweet'].apply(nfx.remove_multiple_spaces)

Data['tweet']=Data['tweet'].apply(nfx.remove_special_characters)
Data['tweet']=Data['tweet'].apply(nfx.remove_bad_quotes)
```

```
Data['tweet']
```

```
import keras
from keras.preprocessing.text import Tokenizer
tok1 = Tokenizer(char_level=False, filters ='!"$%&@()*+,-./:;”“<=>?[\\]^_`{|}~\t\n', lower = True)
```

```
tok1.fit_on_texts(Data['tweet'])
words=len(tok1.word_counts)
words
```

```
word_list=[tok1.word_index]
word_list
```

```
encoded=tok1.texts_to_sequences(Data['tweet'])
print(Data['tweet'][0])
encoded[0]
```

```
import numpy as np
from keras.utils.np_utils import to_categorical
classes_list = ["real","fake"]
label_index = Data['label'].apply(classes_list.index)
label1 = np.asarray(label_index)
label = to_categorical(np.asarray(label1))
label
```

```
##         Padding encoded sequence of words
from keras.preprocessing import sequence
max_length=25
padd = sequence.pad_sequences(encoded, maxlen=max_length, padding='post')
padd
```

#CNN

```
# importing libraries for creating model
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten, Embedding,MaxPool1D
from keras.layers import Conv1D, GlobalMaxPooling1D, MaxPooling1D
```

```
model = Sequential()
model.add(Embedding(input_dim = words + 1, output_dim = 100,input_length=max_length))
model.add(Conv1D(filters =32,kernel_size=2, activation='relu'))
model.add(MaxPooling1D(3))
model.add(Flatten())
model.add(Dense(32, activation = 'relu'))
#model.add(Dense(16, activation = 'relu'))
model.add(Dense(2, activation = 'sigmoid'))
model.summary()
```

```
## compile

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])
```

```
model.fit(padd,label,epochs=20,verbose=1,batch_size=32)
```
