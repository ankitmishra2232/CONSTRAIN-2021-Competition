```
Data=pd.read_excel('drive/My Drive/Ankit/october2020 codalab/hindi/constraint_Hindi_Train.xlsx')
Data.dropna(inplace = True)
Data.head(10)
```

```
#remove emoji

import re

def deEmojify(txt):
    regrex_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"             # emoticons
        u"\U0001F300-\U0001F5FF"             # symbols & pictographs
        u"\U0001F680-\U0001F6FF"             # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"             # flags (iOS)
        u"\U0001f92a"
        u"\U0001f9b0"
        u"\U0001f92b"
        u"\U0001f928"
        u"\U0001f92d"
                           "]+", flags = re.UNICODE)
    return regrex_pattern.sub(r'',txt)

Data['Post']=Data['Post'].apply(lambda x:deEmojify(x))
```

```
# remove URL and numbers
import string
def removeURL(txt):
  url=re.compile(r"https?://\S+")
  return url.sub(r"",txt)

Data['Post']=Data['Post'].apply(lambda x:removeURL(x))
Data['Post']=Data['Post'].str.replace(r'\d+','')
Data.head()
```

```
def remove(txt):
  url=re.compile(r"@#\n\n")
  return url.sub(r"",txt)
```

```
Data['Post']=Data['Post'].apply(lambda x:remove(x))
```

```
x=Data['Post'].to_string()
x.replace(r'\\n@#','')
```
